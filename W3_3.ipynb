<H2>Importing the content from wiki and checking to see the title</H2>

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(url,'lxml')
soup.title.string
```

<H2>Loading the table data</H2>

```python
ld_table=soup.find('table',{'class':'wikitable'})
ld_table
```

<H2>Looping through the rows and columns of the Wiki table into temp dataframes</H2>

```python
A=[]
B=[]
C=[]

for row in ld_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))

        
```

```python
df=pd.DataFrame(columns=['Postcode','Borough','Neighbourhood'])
df['Postcode']=A
df['Borough']=B
df['Neighbourhood']=C
df
```

```python
df=df.replace({'\\n':''}, regex=True)
df['Borough']=df['Borough'].replace({'Not assigned':''})
df['Borough'].replace('', np.nan, inplace=True)
df.dropna(subset=['Borough'], inplace=True)
df['Neighbourhood']=df['Neighbourhood'].replace({'Not assigned':df['Borough']})
df
```

```python
df=df.groupby('Postcode').agg({'Borough':'first','Neighbourhood':', '.join}).reset_index()
df
```
